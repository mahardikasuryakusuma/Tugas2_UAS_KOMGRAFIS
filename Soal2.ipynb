```python
# Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Mengambil data dari file CSV
data = pd.read_csv('data.csv')
# Menampilkan data
data
```

```python
# Mengelompokan data dari file CSV
labels = data.columns.tolist()
asimilasi = data['Asimilasi']
observasi = data['Observasi']
non_asimilasi = data['Non Asimilasi']

# Mengatur ukuran canvas
plt.figure(figsize=(15, 5))

# Mengatur ketebalan line pada line chart
plt.plot(asimilasi, marker='o', label='Asimilasi', linewidth=3, color="#22A699")
plt.plot(observasi, marker='o', label='Observasi', linewidth=3, color="#FFA559")
plt.plot(non_asimilasi, marker='o', label='Non Asimilasi', linewidth=3, color="#FF6000")

# Memberi label keterangan dan judul
plt.xlabel('Waktu')
plt.ylabel('Celcius')
plt.title('Suhu Permukaan', size = 16)

# Mengatur range value pada sumbu Y
plt.yticks(np.arange(20, 28, 1))

# Mengatur value label pada sumbu X
waktu_awal = datetime(2013, 7, 24)
x_ticks = np.arange(0, len(data))
x_label = []
for i in range(len(data)):
    days_delta = i // 8
    time = waktu_awal + timedelta(days=days_delta)
    hour = (i % 8) * 3
    label = time.strftime('%d %b %Y') + ' ' + str(hour).zfill(2) + ':00 UTC'
    x_label.append(label)

plt.xticks(x_ticks, x_label, rotation=45)

# Memberi garis horizontal pada canvas
plt.grid(axis='y')

# Mengatur posisi legenda
plt.legend(loc='lower left')
# Menampilkan multiline chart
plt.show()
```
