```python
# Import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
%matplotlib inline

# Mengambil data dari file CSV
data_cuaca = pd.read_csv('Komgraf.csv')
# Menampilkan data file CSV
data_cuaca
```

```python
# Melakukan pengelompokan data sesuai tahun
data_cuaca_1981_1990 = data_cuaca['1981-1990']
data_cuaca_1991_2000 = data_cuaca['1991-2000']
data_cuaca_2001_2010 = data_cuaca['2001-2010']

# Mengatur lebar bar
width = 0.22
# Mengatur ukuran kanvas
fig, ax = plt.subplots(figsize=(17, 7))

# Mencetak chart bar berdasar pengelompkan data yang sudah dilakukan
x = np.arange(len(data_cuaca))
tahun1 = ax.bar(x - width, data_cuaca_1981_1990, width, label='1981-1990', color='#E83845', edgecolor='black')
tahun2 = ax.bar(x, data_cuaca_1991_2000, width, label='1991-2000', color='#746AB0',edgecolor='black')
tahun3 = ax.bar(x + width, data_cuaca_2001_2010, width, label='2001-2010', color='#288BA8',edgecolor='black')

# Memberi grid/garis-garis horizontal pada kanvas
ax.yaxis.set_major_locator(MultipleLocator(100))
ax.yaxis.set_minor_locator(MultipleLocator(20))
ax.set_axisbelow(True)
ax.grid(axis='y', linestyle='--', which='major', color='gray')
ax.grid(axis='y', linestyle=':', which='minor', color='gray')

# Memberi label pada sumbu X dan Y
ax.set_ylabel('Curah Hujan (mm)', size=14)
ax.set_xticks(np.arange(len(data_cuaca)))
ax.set_xlabel('Bulan',size=14)
ax.set_xticklabels(data_cuaca.index + 1, size=12)
# Mengatur posisi legenda
plt.legend(fontsize=14, loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=3)
# Mencetak bar
plt.show()
```
